```python
#dropping irrelevant columns
df=df.drop(columns=["Unnamed: 0"])
```

```python
#splitting data into x and y data
# x dataset containing all columns except the last one. It's a way to select a subset of the original DataFrame, excluding the last column.
x=df.iloc[:, 0:-1]
```

```python
x.head(5)
```

```python
#y dataset conataining only the last column
y=df.iloc[:,-1]
```

```python
y
```

```python
#training and testing x and y data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
#converting the training and testing dataset to intger
x_train=x_train.astype(int)
y_train=y_train.astype(int)
x_test=x_test.astype(int)
y_test=y_test.astype(int)
```

```python
#using the mean and standard deviation computed from the training data.
Sc=StandardScaler()
x_train_scaled=Sc.fit_transform(x_train)
x_test_scaled=Sc.fit_transform(x_test)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test_scaled)
```

```python
#using r2 square method to quantify how well the predictions are
r2_score(y_test,y_pred)
```

```python
#analyzing the data by heatmap
plt.hexbin(y_test, y_pred, gridsize=25, cmap='viridis', bins='log', mincnt=1)
plt.colorbar(label='log10(N)')
```
